```python
sql = """
SELECT GameID, GameUserClass AS User_Class, GameOpponentClass AS Opp_Class, Victory FROM game
WHERE GameOpponent != "The Innkeeper"
"""

classes = pd.read_sql(sql, con=engine)
classes.head()
```

```python
classes.Victory = classes.Victory.map(ord)
```

```python
def class_converter(df):
    for x in df:
        if x == 'Jaina Proudmoore' or 'Medivh' or 'Khadgar':
            x.replace('Mage')
        elif x == "Thrall":
            x = 'Shaman'
        elif x == 'Guldan':
            x = 'Warlock'
        elif x == 'Rexxar' or 'Alleria Windrunner':
            x = 'Hunter'
        elif x == 'Anduin Wrynn':
            x = 'Priest'
        elif x == 'Uther Lightbringer' or 'Lady Liadrin':
            x = 'Paladin'
        elif x == 'Garrosh Hellscream' or 'Magni Bronzebeard':
            x = 'Warrior'
        elif x == 'Valeera Sanguinar':
            x = 'Rogue'
        elif x == 'Malfurion Stormrage':
            x = 'Druid'
            
class_converter(classes.User_Class)
```

```python
classes['Matchup'] = [(str(classes.User_Class[i]) + str(classes.Opp_Class[i])) for i in range(len(classes.User_Class))]
```

```python
sql = """
SELECT game.GameID, game.GameUserName AS User, game.GameUserClass AS UserClass, game.GameOpponent AS Enemy, 
game.GameOpponentClass AS EnemyClass, game.Victory, game.Coin, turn.TurnID, turn.TurnNumber, cardturn.MinionName, 
cardturn.Health, cardturn.Friendly, cardturn.Cost FROM game
JOIN turn
ON game.GameID = turn.GameID
JOIN cardturn
ON cardturn.TurnID = turn.TurnID
WHERE (game.GameOpponent != "The Innkeeper") AND (cardturn.Friendly = 1)
"""

test = pd.read_sql(sql, con=engine)
test
```

```python
test.TurnNumber = test.TurnNumber.apply(int)
```

```python
test.Victory = test.Victory.map(ord)
test.Coin = test.Coin.map(ord)
test.Friendly = test.Friendly.map(ord)
test.head(15)
```

```python
test['Mana'] = [test.TurnNumber[i] for i in range(len(test.TurnNumber))]
```

```python
test.Mana = test.Mana.map(lambda x: 10 if x > 10 else x)
```

```python
sql = """
  SELECT 
  g.GameID,
  g.GameUserName,
  ct.TurnNumber,
  SUM(Cost) AS TurnCost,
  ct.Friendly
FROM Game AS g
INNER JOIN Turn AS t ON g.GameID = t.GameID
INNER JOIN CardTurn AS ct  ON t.TurnID = ct.TurnID
WHERE ct.MinionName <> "Guldan" AND ct.MinionName <> "Uther Lightbringer"  AND ct.MinionName <> "Garrosh Hellscream" 
AND ct.MinionName <> "Malfurion Stormrage"
AND ct.MinionName <> "Rexxar" AND ct.MinionName <> "Jaina Proudmoore" AND ct.MinionName <> "Anduin Wrynn" AND ct.MinionName <> "Valeera Sanguinar" 
AND ct.MinionName <> "Thrall" AND Friendly = 0
GROUP BY ct.TurnNumber
Order By ct.TurnNumber
"""

mana = pd.read_sql(sql, con=engine)
mana
```
