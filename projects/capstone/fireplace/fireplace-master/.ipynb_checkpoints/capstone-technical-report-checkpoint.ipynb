## Problem Statement and Initial Questions

The goal of the project is to accurately predict each player's chance of winning in Hearthstone, an online trading card game. The idea behind this is two-fold: to help players improve their play by showing which plays increase or decrease their chances of winning, and to help spectators (particularly new players) gain a better understanding of the game's events, similar to how poker displays each players chance of winning. The theory is that despite Hearthstone’s infamous randomness, a player’s chance of winning can be accurately predicted each turn based on a number of features, such as mana efficiency, cards mulliganed, and board state at each turn.

There are many questions that arise from this idea: What's the best way to gather the necessary amounts of data in order to build a model? How much data is even needed? If a lot of data is required, does this become a big data problem? How can we tackle big data? Are there any computational limitations to keep in mind? Once the data is gathered, what kind of model should be used based on the type of data?

In retrospect, this was a very ambitious project. However, through a potent combination of stubbornness and naivety, much progress was made on the project as indicated below. 

## Generating/Gathering the Data

The data used to train and test the model below was generated from an open source Hearthstone simulator called Fireplace. You can find the simulator [here](https://github.com/jleclanche/fireplace). However, the scripts had to be edited significantly for a few reasons:
- The simulations weren't truly random, so the code was edited to implement random factors in each and every simulation.
- A logger was built to gather the data from the simulations (see below), and code had to edited and added in order to wrap the simulation around said logger.
- A few bugs were present in the code that needed to be fixed.

Once the simulations were working as intended, the data generated from the simulations needed to be logged in some fashion. Dave helped build a script that uses a web server to log the games using a unique ID generator as an ID for each game simulated. Initially, a logger was written that didn't use a web server to generate unique ID's but just incremented them instead. This turned out to be problematic as the game IDs of simulations' that were run simultaenously in a threaded process were overlapping, causing games to have the incorrect game ID. This was fixed once ran through a web server as the logger would force each game to have a unique ID rather than just assigning an incremental variable. As the simulations were run, they were stored in a local MySQL Database, which would then be exported later for cleaning and munging. 

The eventual model for this project should ideally take millions of games, and as such running the logger by itself would require far too much time. Thus, both Dave and Kiefer helped build a threaded process that can run several instances of the logger, drastically increasing the rate at which games were collected. Furthermore, the threaded process was run on multiple AWS EC2 instances, as the threaded process was too computationally intensive for a local machine. Given the time constraint, only two days of simulations were run, resulting in over 35,000 games and 6,000,000 rows. Unfortunately this isn't nearly enough to have an ideal model, but it's a good enough starting place. 

## Cleaning/Munging the Data

After the above data was generated and gathered in a SQL database, it was then exported to multiple `.csv` files for each EC2 instance. The data here needed to be in a format that our eventual model would recognize best. This was initially done using the `pandas` package in python, but this task proved to be the most challenging problem in the entire project, and was much more easily done in `R`. Thus, the `.csv` files were imported to `R` where they were cleaned and munged by Kiefer. The data here is transformed from long to wide format, converting every unique event's value to a column, where it is then populated by either a 0 if it didn't happen or a 1 if it did happen. There are cases where a card can appear in a game multiple times, and so these columns can be populated with 2, 3, etc as well. The resulting data is several GBs big with over 9,000 columns. As a result, the data had to be divied up into multiple batches to make it computationally reasonable to train a model on. However, as described below under the **Modelling** section, this causes a problem with most `scikit-learn` packages as we can't fit models on multiple datasets at once. Luckily for us, `scikit-learn` has a few modelling packages that have a `partial_fit` method that can iteratively fit a model on multiple datasets.

## Modelling the Data

Below is the meat of the project, in which we finally choose a model and train it on the above data. The model chosen for this project is the Multinomial Naive Bayes Classifier as it satisfies multiple criteria that we need to fulfill, such as the `partial_fit` method (described above) and its ability to classify discrete features. 

```python
import pandas as pd
import numpy as np
import cPickle
import gzip
import time
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier
from sklearn.cross_validation import cross_val_score, StratifiedKFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.naive_bayes import MultinomialNB, GaussianNB
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_curve, roc_auc_score, auc
from sklearn.tree import DecisionTreeClassifier
from sklearn.grid_search import GridSearchCV
%matplotlib inline
```

```python
path = '../../../../../fireplace_datasets/widened_piece_csvs/wide%d.csv'
```

Before we iterate through all batches of data and perform `partial_fits` on them, we first need to instantiate the first dataframe. This is because the `partial_fit` method needs to be explicitly told what classes to expect before it runs on any dataset. To do this, we pass the `classes` parameter and give it [0, 1] as those are the only possible values for our target variable (whether or not a player won or lost). 

```python
df1 = pd.read_csv(path % 1)
```

```python
df1.shape
```

Here we instantiate the Multinomial Naive Bayes Classifier. There are three total parameters for MNB: `alpha`, `fit_prior`, and `class_prior`, however we won't be using any of them as they don't significantly affect our model.

```python
# using Multinomial Naive Bayes as our first model
mnb = MultinomialNB()
```

Below we create a list of columns to make certain that the order of columns for each batch is the same across all batches. Also we create our predictors and exclude irrelevant columns as well as our target variable.

```python
columns = df1.columns
# should I include game_turn in preds? and remove player_event_num?
preds = [col for col in columns if col not in ['event_id', 'player', 'game_id', 'game_event_id', 'player_won', 'game_turn_max']]
```

Now, we need to remove the last two turns from each game as those turns are direct indicators of a player winning or losing a game. We don't want our model to figure out how to cheat the system.

```python
max_game_turn = df1.groupby(['game_id'])['game_turn'].max()
df1 = df1.join(max_game_turn, how = 'right', on = 'game_id', rsuffix = '_max')
df1 = df1[(df1.game_turn + 1) < df1.game_turn_max]
```

Next we fill all `NaN` values with 0, instantiate our target and predictor columns, and `partial_fit` the model onto `df1`. We pass two parameters in for the `partial_fit` method, `classes` and `sample_weight`. `classes`, as already described above, states the possible values for our target variable, and `sample_weight` tells the model how much importance to put on the batch of data being fitted. This parameter is very important, which will be demonstrated further below. A `sample_weight` of 1 is the default value. 

```python
df1.fillna(value = 0, inplace = True)
y = df1.player_won
X = df1[preds]
weight = 1
mnb.partial_fit(X, y, classes = [0, 1], sample_weight = weight)
scores = cross_val_score(mnb, X, y, cv = 11)
print scores
print scores.mean()
```

Next we'll create several functions that we will eventually iterate through. This first function, `csv_loader`, simply loads a dataframe based on where we are in our iteration and fills all `NaN` values with 0 (otherwise our iteration will return an error when attempting to `partial_fit` the model). 

```python
def csv_loader(i):
    df = pd.read_csv(path % i)
    df.fillna(0, inplace=True)
    return df
```

The next function, `partial_fit_mnb`, takes a model, dataframe and a number that represents the amount of rows of the previous dataset. 

First, the function calculates the weight to pass to the `sample_weight` parameter of the `partial_fit` method of our model. This essentially tells model how much emphasis to apply to the current batch of data relative to the amount of data we've fitted previously. For example, a batch of data that's fit two interations in will have much more impact on our model than a batch of data 100 iterations in. 

Next, the function instantiates the target variable (Y) and the predictor variables (X) that we then pass to the `partial_fit` method of our model. Along with X and Y, we pass it the previously calculated `sample_weight` value as well as the `classes` parameter to explicitly tell it the possible values of our target variable (in this case, whether a player won or not).

Finally, the function returns the fitted model so that we can compute various metric scores below. 

```python
def partial_fit_mnb(mod, df, priorsize):
    weight = df.shape[0]/float(priorsize + df.shape[0])
    Y = df.player_won.values
    X = df[[c_ for c_ in df.columns if not c_ == 'player_won']].values
    print 'win_pct:', np.mean(Y)
    mod.partial_fit(X, Y, sample_weight=weight, classes=[1,0])
    return mod
```

Our final function, `test_acc_mnb`, takes our model from above as well as a dataframe of the next three batches of data. This is to test the model on multiple datasets that it hasn't encountered yet. To do this, we first instantiate Y and X (target and predictor variables) and print the accuracy of the model's predictions via the `score` method. We also want to return the score, as we'll be compiling these accuracies in a list and plotting them to visualize the course of the model's accuracy as it fits on more and more batches of data. 

```python
def test_acc_mnb(mod, test):
    Y = test.player_won.values
    X = test[[c for c in test.columns if not c == 'player_won']].values    
    
    score = mod.score(X, Y)
    
    print 'test acc score:', score
    return score
```

The final step before we start iterating through all of our batches of data is to instantiate various variables that we'll be using through the following for loop. Below is a brief description of each variable.

- `iterscores`: A list that we'll be appending the accuracy scores described above to help visualize the data's performance.
- `prior_size`: A value that represents the size of the previous batch of data.
- `corder`: A variable that is assigned the columns of a dataframe which are then ordered in a specific way to ensure that the order of columns for all batches is the same.
- `itermnb`: Our model, no parameters passed.
- `remove_cols`: A list of columns that we want to remove from all datasets.

```python
iterscores = []
prior_size = 0
corder = None
itermnb = MultinomialNB()

remove_cols = ['game_id','event_id','player']
```

Now, we finally iterate through every batch and perform various operations on them. Since we want to test the model on the next four batches, we must begin by loading the first four batches of data. Then, we merge these into one dataframe, called `cdf`, and call the `partial_fit_mnb` function (described above), passing our model and `cdf` into it. Next, we assign the rows of `cdf` to `prior_size` so that we may use this to determine how much weight to apply to the next `partial_fit`. Finally, we delete `cdf` to ensure our computer doesn't explode. 

This process changes slightly once we've fit the first four batches on the model. When we get to the second iteration (i = 5), we fit the model on the 5th dataset using the `partial_fit_mnb` function and merge the next three batches into one dataframe (named `testdf`). We then test the model on these three batches using `test_acc_mnb`, assigning the resulting accuracy score to `current_score`. Then, we append `current_score` to the earlier created `iterscores` list and delete `testdf`, ensuring once again the health of our machine. Finally, we assign the rows of the first dataframe that we fit the model on to `prior_size`, again so we can use this value to determine how much emphasis we place on the next `partial_fit`. 

With the above situations initalized, we can finally begin iterating through each batch of data. We do this by setting the dataframes equal to the next dataframe, overwriting the previously saved dataset. This doesn't work for the 4th next dataset (as we're only loading 4 batches of data at a time) so we simply set it equal to the 3rd next dataset. Then, similar to the previous step, we create `testdf`, test the model on it, pring and store the resulting accuracy score, and delete `testdf`. 

One final note: the for loop can run into what seems to be a memory related error. It's not clear what causes this error, but research and trial and error have determined that explicitly telling the system to wait for a short amount of time (hence the `time.sleep(0.1)` fixes this error. Again, it is unclear what causes this error, but for the sake of time (and sanity) this solution is sufficient for now.

```python
for i in range(4, 113):
    print '---------------------------------------------------'
    print 'Iteration:', i
    if i == 4:
        a = csv_loader(i)
        b = csv_loader(i-1)
        c = csv_loader(i-2)
        d = csv_loader(i-3)
        
        print 'unique gids:', len(d.game_id.unique())
        
        corder = a.columns
        corder = [c_ for c_ in corder if c_ not in remove_cols]
        
        a = a[corder]
        b = b[corder]
        c = c[corder]
        d = d[corder]
        
        cdf = pd.concat([a,b,c,d], axis=0)
        itermnb = partial_fit_mnb(itermnb, cdf, prior_size)
        
        prior_size = cdf.shape[0]
        del cdf
        
    elif i == 5:
        
        a = csv_loader(i)
        b = csv_loader(i+1)
        c = csv_loader(i+2)
        d = csv_loader(i+3)
        
        print 'unique gids:', len(d.game_id.unique())
        
        a = a[corder]
        b = b[corder]
        c = c[corder]
        d = d[corder]
        
        itermnb = partial_fit_mnb(itermnb, a, prior_size)
        time.sleep(0.1)
        
        testdf = pd.concat([b,c,d], axis=0)
        
        current_score = test_acc_mnb(itermnb, testdf)
        
        iterscores.append(current_score)
        
        del testdf
        
        prior_size = prior_size + a.shape[0]
        
    elif i > 4:
        a = b
        b = c
        c = d
        d = csv_loader(i+3)
        
        print 'unique gids:', len(d.game_id.unique())
        
        d = d[corder]
        
        itermnb = partial_fit_mnb(itermnb, a, prior_size)
        time.sleep(0.1)
        
        testdf = pd.concat([b,c,d], axis=0)
        
        current_score = test_acc_mnb(itermnb, testdf)
        
        iterscores.append(current_score)
        
        del testdf
        
        prior_size = prior_size + a.shape[0]
```

Below is a visual representation of the accuracy scores through all iterations. Notice that the accuracies get less noisy the more data is fit to the model.

```python
sns.set_style('white')

plt.figure(figsize=(12, 7))

x = np.arange(5, 113)
y = iterscores

plt.plot(x, y, c='steelblue', lw=3)
plt.axhline(y=0.515, ls='dashed', c='grey', lw=2, alpha=0.7)

plt.show()
```

Here we simply fill the concatenated dataframe's `NaN` values with 0 and instantiate the target variable, Y, and the predictor variable, X. 

```python
testdf.fillna(0, inplace=True)
Y = testdf.player_won.values
X = testdf[[c for c in testdf.columns if not c == 'player_won']].values
```

Now we can use our model to predict whether or not a player won or lost as well as the model's probablities of those predictions. We're using `itermnb` which has been fitted on the previous 112 batches of data.

```python
mnb_preds = itermnb.predict(X)
mnb_pp = itermnb.predict_proba(X)[:, 1]
```

Below, we have our classification report for our predictions. The `precision` score is the percentage of how many of our predictions were correct, whereas the `recall` score is the percentage of how many true values we correctly predicted. The `f1-score` is simply the harmonic mean of both `precision` and `recall`. 

```python
print(classification_report(Y, mnb_preds))
```

Next is the confusion matrix of our predictions versus the true values, which simply contains the counts of both correct and incorrect predicted values. Something to look for here is if the model is almost exclusively predicting one value. Luckily for us we have a good mix of predictions.

```python
print pd.crosstab(Y, mnb_preds, rownames=['True'], colnames=['Predicted'], margins=True)
```

One of the last things we'll be doing is visualizing our model's AUC score with the ROC curve. Below we're computing the false positive rate, true positive rate, and the thresholds that the ROC curve sets to decide in which class to put our predicted values in. We're also computing the AUC score using the `roc_auc_score` function, which we'll be plotting alongside the `fpr` and `tpr` values.

```python
fpr, tpr, thresholds = roc_curve(Y, mnb_pp, pos_label=1)
mnb_roc_auc = roc_auc_score(Y, mnb_pp)
print mnb_roc_auc
```

Using matplotlib, we'll plot the ROC curve below. The blue line represents the models AUC score and the dashed red line represents the baseline (i.e. if we just randomly guessed). Here we have an AUC score of 0.66, which is quite reasonable considering the low amount of games we have and the randomness of said games. 

```python
plt.figure(figsize=(8, 8))
plt.title('Receiver Operating Characteristic for Multinomial Naive Bayes Classifier')
plt.plot(fpr, tpr, 'b', label='AUC = %0.2f'% mnb_roc_auc)
plt.legend(loc='lower right')
plt.plot([0,1], [0,1], 'r--')
plt.xlim([0, 1.0])
plt.ylim([0, 1.0])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
