```python
# converting unicode values to strings
df = df.astype(str)

# cleaning the event_values column
df.event_value.replace(to_replace = ['<', '>', '\(', '\)', 'Minion ', 'Spell ', 'Weapon ', 'Secret '],
                       value = '', regex = True, inplace = True)
```

```python
df['timestamp'] = pd.to_datetime(df.timestamp, format='%Y-%m-%d %H:%M:%S:%f')
```

```python
df.sort_values(['timestamp'], inplace=True, ascending=True)
```

```python
df.player = df.player.map(lambda x: x.lower())
```

```python
def create_player_turn(subset):
    unique_turns = subset[subset.event_key == 'turn_begins'].shape[0]
    subset['player_turn'] = 0.
    subset.ix[subset.event_key == 'turn_begins', 'player_turn'] = np.ones(unique_turns)
    subset['player_turn'] = np.cumsum(subset.player_turn)
    return subset

df = df.groupby(['game_id','player']).apply(create_player_turn).reset_index(drop=True)
```

```python
def create_overall_turn(subset):
    unique_turns = subset[subset.event_key == 'turn_begins'].shape[0]
    subset['overall_turn'] = 0.
    subset.ix[subset.event_key == 'turn_begins', 'overall_turn'] = np.ones(unique_turns)
    subset['overall_turn'] = np.cumsum(subset.overall_turn)
    return subset
    
df = df.groupby(['game_id']).apply(create_overall_turn).reset_index(drop=True)

df = df[~(df.event_key.isin(['turn_begins','turn_end','first_player']))]
```

```python
# applying the above function to the event_values that are lists
list_events = df[df.event_key.isin(['deck', 'cards_mulliganed', 'cards_kept', 'discover_choices',
                                    'hand', 'mulligan_choices', 'board_state', 'timestamp',
                                    'game_id', 'player'])]
```

```python
# creating a function to convert the format of the data from long to "longer"

def expand_list_cells(subset):
    if len(subset.event_value.values[0]) == 0:
        subset['event_subkey'] = 'empty'
        subset['event_value'] = 1
        return subset
    else:
        ndf = pd.concat([subset]*len(subset.event_value.values[0]), ignore_index=True)
        ndf['event_subkey'] = subset.event_value.values[0]
        ndf['event_value'] = 1
        return ndf
```

```python
list_events = list_events.groupby(['timestamp', 'game_id', 'player']).apply(expand_list_cells)
list_events.reset_index(drop=True, inplace=True)
```

```python
keys_not_in_mask = ~df.event_key.isin(['deck', 'cards_mulliganed', 'cards_kept', 'discover_choices',
                                       'hand', 'mulligan_choices', 'board_state', 'attack'])

singular_events = df[keys_not_in_mask]

singular_events.event_value.replace(to_replace = [' '], value = '_', regex = True, inplace = True)
```

```python
singular_events['event_subkey'] = singular_events.event_key
singular_events['event_value'] = 1
```

```python
list_events['event_key'] = list_events.event_key.map(lambda x: x.lower())
list_events['event_subkey'] = list_events.event_subkey.map(lambda x: x.lower())
list_events['player'] = list_events.player.map(lambda x: x.lower())

singular_events['player'] = singular_events.player.map(lambda x: x.lower())
singular_events['event_key'] = singular_events.event_key.map(lambda x: x.lower())
singular_events['event_subkey'] = singular_events.event_subkey.map(lambda x: x.lower())
```

```python
expanded = pd.concat([list_events, singular_events], axis=0, ignore_index=True)
```

```python
def add_event_indicator_columns(subset, skipcols=[]):
    if subset.event_key.values[0] not in skipcols:
        subset['event_'+subset.event_key.values[0]] = 1
    return subset

def add_persistent_event_columns(subset, ecols=[]):
    if subset.event_key.values[0] in ecols:
        subset['state_'+subset.event_key.values[0]] = subset.event_value.values.astype(np.float64)
    return subset
```

```python
persistent_event_keys = ['board_state','avg_hand_cost','current_hero_power',
                         'deck_cost','max_mana','num_minions','num_spells',
                         'num_weapons','player_class','deck']

remove_event_keys = ['current_hero_power','player1_class','player2_class']
```

```python
exp_groups = expanded.groupby(['event_key','player','game_id'])
expanded_keys = exp_groups.apply(add_persistent_event_columns, ecols=persistent_event_keys).reset_index(drop=True)

print expanded_keys.columns

for pc in persistent_event_keys:
    expanded_keys['state_'+pc] = expanded_keys['state_'+pc].fillna(method='ffill')
    expanded_keys['state_'+pc] = expanded_keys['state_'+pc].fillna(method='bfill')
    
    
#print expanded_keys.loc[:,persistent_event_keys].fillna(method='ffill', axis=1)
#expanded_keys.loc[:,persistent_event_keys].fillna(method='backfill', inplace=True, axis=0)
```

```python
expanded_keys = expanded_keys[~(expanded_keys.event_key.isin(remove_event_keys+persistent_event_keys))]
```

```python
exp_groups = expanded_keys.groupby(['event_key','game_id','player'])
expanded_keys = exp_groups.apply(add_event_indicator_columns, skipcols=persistent_event_keys).reset_index(drop=True)

for col in ['event_'+c for c in expanded_keys.event_key.unique() if c not in persistent_event_keys]:
    expanded_keys[col] = expanded_keys[col].fillna(0.)
    expanded_keys[col] = expanded_keys[col].fillna(0.)
```

```python
def widen_func(x):
    try:
        return x[0]
    except:
        return 1

indexes = [c for c in expanded.columns if c not in['event_value','event_subkey']]
games_wide = pd.pivot_table(expanded, index=indexes,
                            columns=['event_subkey'],
                            values=['event_value'],
                            aggfunc=widen_func)['event_value'].reset_index()

for c in games_wide.columns:
    games_wide[c] = games_wide[c].fillna(0)
```
