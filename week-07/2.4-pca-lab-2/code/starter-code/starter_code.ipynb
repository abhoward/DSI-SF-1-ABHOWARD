## Step 3: Define the Variables

Next, let's define the x and y variables: Airport is going to be our "x" variable

```python
y = 'airport'
X = airports[[c for c in airports.columns if c != 'airport']]
```

## Step 4: Covariance Matrix

Then, standardize the x variable for analysis

```python
Xn = ((X - X.mean()) / X.std()).values
```

```python
Xn = pd.DataFrame(Xn)
```

Next, create the covariance matrix from the standardized x-values and decompose these values to find the eigenvalues and eigenvectors

```python
corr = Xn.corr().values
corr
```

```python
eigenvalues, eigenvectors = np.linalg.eig(corr)
```

# Step 5: Eigenpairs

Then, check your eigenvalues and eigenvectors:

```python
print eigenvalues
print ''
print eigenvectors
```

To find the principal componants, find the eigenpairs, and sort them from highest to lowest. 

```python
# pair eigenvalues and eigenvectors, then sort biggest to smallest by eigenvalue
value_vector_pairs = [[eigenvalues[i], eigenvectors[:, i]] for i in range(len(eigenvalues))]
value_vector_pairs.sort(reverse = True)
```

```python
# create the projection matrix W
weight_2d_projection = np.hstack((value_vector_pairs[0][1].reshape(eigenvectors.shape[1], 1),
                                 value_vector_pairs[1][1].reshape(eigenvectors.shape[1], 1)))

print weight_2d_projection
```

## Step 6: Explained Variance

Now, calculate the explained variance and the Cumulative explained variance

```python
eigen_total = np.sum([eigenvalues[i] for i in range(len(eigenvalues))])

sum = 0

for i in range(len(eigenvalues)):
    eigenscore = eigenvalues[i] / eigen_total
    print eigenscore
    sum = sum + eigenscore
    print "SUM: ", sum
```

** What does the explained variance tell us?**: Here, we can see that 81.77% of the behavior can be explained by the first two principal components
