```python
import pandas as pd
import time
import datetime
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
wow = pd.read_csv('assets/datasets/wowah_data.csv', encoding = 'utf-8')
```

```python
wow = wow.sample(frac = 0.01, random_state = 1234)
```

```python
wow.head()
```

```python
wow.columns = ['char', 'level', 'race', 'charclass', 'zone', 'guild', 'timestamp']
```

```python
wow.columns
```

```python
wow.head()
```

```python
wow.timestamp =  pd.to_datetime(wow.timestamp, infer_datetime_format = True)
```

```python
last_zones = wow.groupby(['zone', 'char'])['timestamp'].max().reset_index()
last_zones.head()
```

```python
max_lifetime = wow.groupby(['char'])['timestamp'].max().reset_index()
min_lifetime = wow.groupby(['char'])['timestamp'].min().reset_index()
```

```python
max_lifetime.head()
```

```python
min_lifetime.head()
```

```python
avg_lifetime_per_char = [max_lifetime['timestamp'][i] - min_lifetime['timestamp'][i] for i in range(len(max_lifetime['timestamp']))]
```

```python
avg_lifetime_per_char
```

```python
len(avg_lifetime_per_char)
```

```python
avg_ms = pd.Series(avg_lifetime_per_char)
```

```python
avg_ms = avg_ms / pd.Timedelta(microseconds=1)
avg_ms = pd.Timedelta(avg_ms.mean(), unit='us')
print avg_ms
```

```python
average_timedelta = sum(avg_lifetime_per_char, datetime.timedelta(0)) / len(avg_lifetime_per_char)
```

```python
average_timedelta
```

```python
test = pd.DataFrame([last_zones.ix[i, :] for i in range(len(last_zones.timestamp)) if last_zones.timestamp[i].month != 12])

stuff = test.zone.value_counts()
```

```python
stuff
```

```python
stuff = pd.DataFrame(stuff).reset_index()
```

```python
stuff.rename(columns = {'index': 'city', 'zone': 'num_players'}, inplace = True)
```

```python
stuff
```

```python
sns.set(style = 'whitegrid', rc = {'figure.figsize':(12, 70)})
sns.barplot(y = 'city', x = 'num_players', data = stuff)
```
